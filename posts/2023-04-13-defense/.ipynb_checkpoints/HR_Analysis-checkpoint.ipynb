```python
import os
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
path = lambda sub: os.path.join("data", sub)
```

```python
HR_data = pd.read_csv(path("HR.txt"), sep=";")
HR_data['time'] =  pd.to_datetime(HR_data['Phone timestamp'] + "000", format='%Y-%m-%dT%H:%M:%S.%f')
```

```python
ECG_data = pd.read_csv(path("ECG.txt"), sep=";")
ECG_data['time'] =  pd.to_datetime(ECG_data['Phone timestamp'] + "000", format='%Y-%m-%dT%H:%M:%S.%f')
```

```python
ECG_data.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4), tight_layout=True)
ax.plot(HR_data["time"], HR_data["HR [bpm]"])
xformatter = mdates.DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(xformatter)
ax.set_ylabel("Heart Rate [bpm]")
ax.grid(True)
# ax.annotate("Defense started", xy=(datetime(2023, 4, 13, 8, 7, 0), 95))
fig.savefig("HR.png", bbox_inches="tight")
plt.show()
```

```python
with plt.xkcd():
    # smoothed version
    kernel_size = 200
    kernel = np.ones(kernel_size) / kernel_size
    fig, ax = plt.subplots(figsize=(10, 4), tight_layout=True)
    hr = np.convolve(HR_data["HR [bpm]"], kernel, "same")
    ax.plot(HR_data["time"][kernel_size:-kernel_size], hr[kernel_size:-kernel_size])
    xformatter = mdates.DateFormatter('%H:%M')
    ax.xaxis.set_major_formatter(xformatter)
    ax.set_ylabel("Heart Rate")
    ax.set_xlim(right=datetime(2023, 4, 13, 11, 0, 0))
    #ax.grid(True)
    # ax.annotate("Defense started", xy=(datetime(2023, 4, 13, 8, 8, 0), 60))
    ax.annotate(
        "PRESENTATION START",
        xy=(datetime(2023, 4, 13, 8, 9, 0), 85), arrowprops=dict(arrowstyle='->'), xytext=(datetime(2023, 4, 13, 7, 45, 0), 95))
    
    ax.annotate(
        "ALL THE QUESTIONS",
        xy=(datetime(2023, 4, 13, 9, 1, 0), 82.2), arrowprops=dict(arrowstyle='->'), xytext=(datetime(2023, 4, 13, 9, 45, 0), 95), ha="center")
    
    for h, m, hr in [(9, 6, 81.2), (9, 11, 80.8), (9, 24, 80.8), (9, 32, 82.7), (9, 38, 79), (9, 45, 82.1), (9, 52, 80.3), (10, 2, 81.2), (10, 12, 80.8), (10, 21, 79.4), (10, 31, 82.2)]:
        #ax.arrow(x=datetime(2023, 4, 13, h, m, 0), y=hr, dx=3, dy=3, length_includes_head=True)
        ax.annotate(
        " "*len("ALL THE QUESTIONS"),
        xy=(datetime(2023, 4, 13, h, m, 0), hr), arrowprops=dict(arrowstyle='->'), xytext=(datetime(2023, 4, 13, 9, 45, 0), 95), ha="center", )
    
        
    ax.annotate(
        "RESULT",
        xy=(datetime(2023, 4, 13, 10, 50, 0), 90.8), arrowprops=dict(arrowstyle='->'), xytext=(datetime(2023, 4, 13, 10, 45, 0), 95), ha="center")
    
    fig.savefig("HR.png", bbox_inches="tight")
```

```python
with plt.xkcd():
    # Based on "Stove Ownership" from XKCD by Randall Munroe
    # https://xkcd.com/418/
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim([-30, 10])
    data = np.ones(100)
    data[70:] -= np.arange(30)
    ax.annotate(
        'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
        xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))
    ax.plot(data)
    ax.set_xlabel('time')
    ax.set_ylabel('my overall health')
    fig.text(
        0.5, 0.05,
        '"Stove Ownership" from xkcd by Randall Munroe',
        ha='center')
```
